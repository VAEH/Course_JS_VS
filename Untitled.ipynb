## Diffing Files (Diferenciar archivos)
**`diff`** Se utiliza cuando queremos comparar 2 tipos de achivo de cdigo y encontrar alguna diferencia dentro de una funcin y/o alguna parte del cdigo.
Cuando se utiliza, obtenemos solo las lineas que son diferentes entre dos archivos. Es mucho ms facil encontrar la diferencia cuando solo tenemos dos lineas.



Ves los smbolos al principio de cada una de esas lneas? El **smbolo menos** que nos dice que la primera lnea fue eliminada del primer archivo, y el **smbolo mayor** que nos dice que la segunda lnea fue agregada al segundo archivo. En otras palabras, la lnea antigua fue reemplazada por la nueva. En este ejemplo, tenamos una lnea que fue reemplazada por una nueva. Este es un cambio comn al modificar cdigo, pero no la nica posibilidad.

Podemos ver que `diff` divide los cambios en dos secciones separadas. La seccin que comienza con **5c5,6** muestra una lnea en el primer archivo que fue reemplazada por dos lneas diferentes en el segundo archivo. El nmero al principio de esta seccin indica el nmero de lnea en el primer y segundo archivo. **La c** entre los nmeros significa que una lnea fue cambiada.

La seccin que comienza con 11a13,15 muestra tres lneas que son nuevas en el segundo archivo. La ***a*** significa que agrego. Aunque las lineas que agrego son algo extraas referente a dicho orden. 
Para entender que paso, se usa la bandera `dash -u ` para que `diff` muestre las diferencias en otro formato.

Este formato unificado muestra las lineas de cambio junto con algn contexto. usando el signo menos para marcar las lneas que se eliminaron, y el signo ms para marcar las lneas que se agregaron. El contexto extra nos permite saber mejor qu est pasando con el cambio que estamos dificando. 

 Podemos ver que el nuevo archivo en realidad tiene un bloque completamente nuevo si que es parte de una cadena de condicionales que se ve muy similar, y es por eso que con la salida diff que vimos antes, fue un poco confuso qu lneas se haban agregado.

* `wdiff` Resalta las palabras que han cambiado en una archivo en lugar de trabajar linea por linea como lo hace `diff`


### Applying Changes
Cuando se quiere dejar claro un cambio dentro del cdigo, se puede enviar aun compaero un `diff` con el cambio para que puedan ver cmo se ve el cdigo modificado, para eso usamos la siguiente linea:
<br>
> diff-u old_file new_ file > change.diff 

El **signo mayor** redirige la salida del comando diff a un archivo. Con este comando estamos generando un archivo llamado *change.diff* con el contenido del comando `diff -i `. usando la bandera `-u`, incluimos ms context que ayuda a la persona que esta leyendo el archivo a entender lo que est pasando con el cambio. 

El **archivo diff** o **archivo patch**, incluye todos los cambios entre el archivo antiguo y el nuevo, ms el contexto adicional necesario para entender los cambios y aplicar esos cambios de nuevo al archivo original. 

**`patch`**, toma un archivo generado por `diff` y aplica los cambios al archivo original.
